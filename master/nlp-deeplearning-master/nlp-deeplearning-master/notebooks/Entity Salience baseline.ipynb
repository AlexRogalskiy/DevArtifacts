# Entity Salience baseline

The goal of this notebook is to replicate the Entity Salience baseline as described in https://www.cs.cmu.edu/~jdunietz/publications/salience.pdf. This notebook is provided by Kevin Jacobs (https://www.data-blogger.com/).

## New York Times corpus
The authors of the paper only use a subset of the New York Times corpus and only use articles from 2003-2007 and by filtering out articles and summaries that were very short or very long, as well as several special article types (e.g., corrections and letters to the editor). The full labeled dataset includes 110,639 documents with 2,229,728 labeled entities; about 14% are marked as salient. 9,719 documents from 2007 are used as test data and the rest as training.

## New York Times annotations
The New York Times corpus with annotations of salient entities is found on https://github.com/google-research-datasets/nyt-salience.

## Loading the data

The first step is to load the data and to align it with the annotations such that it is inline with the paper.

### Check the folders

In the next section, the folders should be specified. Afterwards, it is checked whether the paths are pointing to the correct folders.

```python
import os

# Define the path to the New York Times corpus (a path to a folder consisting of folders per year)
nyt_folder = '/media/sf_Ubuntu/Thesis/data/nyt'
# The path to the New York Salience annotation files (a folder containing nyt_train and nyt_eval)
nyt_salience_folder = '/media/sf_Ubuntu/Thesis/data/nyt-salience'
# Path for storing data
data_path = '../data/'

# Create the data folder
if not os.path.exists(data_path):
        os.makedirs(data_path)
docs_path = os.path.join(data_path, 'df_docs.csv')

# The years used in the paper
years = list(range(2003, 2007 + 1))

# Check whether these folders exist
for year in years:
    year_path = os.path.join(nyt_folder, str(year))
    assert os.path.exists(year_path), 'Path not found: %s' % year_path
    if len(os.listdir(year_path)) == 0:
        raise Exception('No archives found in %s' % year_path)
    
# The annotated train and test files
nyt_train = os.path.join(nyt_salience_folder, 'nyt-train')
nyt_test = os.path.join(nyt_salience_folder, 'nyt-eval')

# Check whether these files exist
assert os.path.exists(nyt_train), 'The annotated train file does not exist: %s' % nyt_train
assert os.path.exists(nyt_test), 'The annotated test file does not exist: %s' % nyt_test

print('All files are in place!')
```

### Load files into memory

The next step is to load the files into memory such that we can work easily with it.

#### Load NYT annotation into memory

The first step is to load all the salient entity annotations into memory.

```python
# My favorite data-loading library
import pandas as pd

def load_annotation(path):
    """
    Loads an annotation file (nyt_train, nyt_eval) into memory.
    """
    doc_id = None
    
    # Document dataframe columns and rows initialization
    df_docs_columns = ['doc_id', 'title']
    df_docs_rows = []
    
    # Entity dataframe columns and rows initialization
    df_entities_columns = ['doc_id', 'e_id', 'is_salient', 'mention_count', 'first_mention', 'byte_offset_start_position_first_mention', 'byte_offset_end_position_first_mention', 'MID']
    df_entities_rows = []
    
    with open(path, 'rb') as input_file:
        lines = input_file.readlines()
        
    for line_number, line in enumerate(lines):
        # Clean the line
        line = line.strip()
        # Find the fields in the line
        fields = line.split("\t")
        # There are two types of lines: those containing entities and lines containing document descriptors
        if len(fields) == 2:
            # It's a document descriptor
            df_docs_rows.append(fields)
            doc_id = fields[0]
        elif len(fields) == 7:
            # It's a entity descriptor
            df_entities_rows.append([doc_id] + fields)
        elif len(fields) == 1:
            # It's an empty line
            pass
        else:
            # Something weird happened
            raise Exception("Unknown line type detected: %s (%d field(s))" % (line, len(fields)))
    
    # Create the document dataframe
    df_docs = pd.DataFrame(df_docs_rows, columns=df_docs_columns)
    
    # Create the entity dataframe
    df_entities = pd.DataFrame(df_entities_rows, columns=df_entities_columns)
    df_entities.doc_id = df_entities.doc_id.astype(int)
    
    return df_docs, df_entities
        
print('Loading %s...' % nyt_train)
df_train_docs, df_train_entities = load_annotation(nyt_train)
print('Loading %s...' % nyt_test)
df_test_docs, df_test_entities = load_annotation(nyt_test)
print('The NYT annotations are loaded.')
```

#### Load NYT corpus in memory

Now load the used files of the NYT corpus into memory.

```python
import xml.etree.ElementTree as ET
import re
import tarfile

if os.path.exists(docs_path):
    # Read the CSV
    df_docs = pd.read_csv(docs_path)
else:
    # Initialize the dataframe
    columns = ['doc_id', 'title', 'abstract', 'content']
    df_docs = pd.DataFrame(columns=columns)

    # Find all valid document identifiers (that are all identifiers from the annotation files)
    valid_doc_ids = {key: '' for key in df_train_docs.doc_id.values.tolist() + df_test_docs.doc_id.values.tolist()}

    # Loop through all directories
    for year in years:
        year_path = os.path.join(nyt_folder, str(year))
        print('Loading %s...' % (year_path))

        # Loop through all archives
        for archive in [archive for archive in os.listdir(year_path) if archive.endswith('.tgz')]:
            path = os.path.join(year_path, archive)
            print('Loading %s...' % path)
            tgz_file = tarfile.open(path)
            # And also loop through all files found within the archive
            members = [member for member in tgz_file.getmembers() if member.isfile()]
            for member in members:
                member_file = tgz_file.extractfile(member)
                lines = member_file.readlines()
                # Find the document ID
                doc_id_line = [line for line in lines if line.strip().startswith('<doc-id ')][0]
                doc_id = doc_id_line.split('"')[1]
                # Check whether this document identifier was found in the annotation files
                if doc_id in valid_doc_ids:
                    content = "\n".join(lines)
                    # Parse the content
                    doc = ET.fromstring(content)
                    # Find title, content and abstract in the parsed XML string
                    title = ''.join(doc.findall('.//title')[0].itertext()).strip()
                    body = ''.join(doc.findall('.//body.content')[0].itertext()).strip()
                    abstract = ''.join(doc.findall('.//abstract')[0].itertext()).strip()
                    # Add a new row
                    df_doc = pd.DataFrame([[doc_id, title, abstract, body]], columns=columns)
                    df_docs = df_docs.append(df_doc)
    
    # Save it
    df_docs.to_csv(docs_path, encoding='utf-8')

# Show it
df_docs
```

### Replicate the baseline

Now we have the data in-place, we can try to replicate the results of the baseline.

#### Positional baseline
Binary feature indicating that an entity is mentioned in the first sentence.

```python
import re
from sklearn.metrics import classification_report
from nltk.tokenize import sent_tokenize

# Select a few samples (the sentence tokenizer takes a long time, therefore it takes a long time to evaluate all samples)
df_samples = df_test_docs.sample(100).drop('title', axis=1)
df_samples = df_docs[df_docs.doc_id.isin(df_samples.doc_id)]

# The predictions and the ground truth
y_preds = []
y_trues = []

# Find all entities according to the samples
df_entities = df_test_entities[df_test_entities.doc_id.isin(df_samples.doc_id)]
df_samples = pd.merge(df_entities, df_docs)

# Loop through all the samples (that is, loop through all the entities)
for index, sample in df_samples.iterrows():
    # Find all sentences
    sentences = sent_tokenize(sample.content.decode('utf-8'))
    # Select the first sentence
    first_sentence = sentences[0].encode('utf-8')
    # The positional baseline looks at whether the entity is mentioned in the first sentence
    y_pred = sample.first_mention in first_sentence
    # Check whether the sample is salient
    y_true = sample.is_salient == '1'
    # And add the score to the score lists
    y_preds.append(y_pred)
    y_trues.append(y_true)
        
# Display the classification report
print(classification_report(y_trues, y_preds))
```

Not bad; the metrics for salient entities are actually pretty close to the metrics in the paper.
